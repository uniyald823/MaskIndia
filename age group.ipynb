```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("C:/Users/Drishya/Desktop/major_project/india_age.csv")  
df = pd.DataFrame(data)  
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
plt.figure(figsize=(20,5))
plt.bar(X, Y, color='r')
plt.title("Female Deaths According to Age Group")
plt.xlabel("Age group")
plt.ylabel("Number of Deaths (in Lakh)")  
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
# Show the plot
plt.show()
```
